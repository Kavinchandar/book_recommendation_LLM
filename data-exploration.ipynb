```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books = pd.read_csv(str(path)+"/books.csv")
books
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax = ax)

plt.xlabel('Columns')
plt.ylabel('Missing Values')

plt.show()
```

```python
import numpy as np

#adding new columns
books['missing_description'] = np.where(books['description'].isna(), 1, 0)
books['age_of_book'] = 2025 - books['published_year']
```

```python
columns_of_interest = ["num_pages", "age_of_book", "average_rating", "missing_description"]

correlation_matrix = books[columns_of_interest].corr(method='spearman')

sns.heatmap(correlation_matrix, annot=True)
```

```python
missing_books = books[(books["description"].isna()) |
       (books["num_pages"].isna()) |
       (books["average_rating"].isna()) |
       (books["published_year"].isna()) |
       (books["description"].isna()) ]

missing_books
```

```python
valid_books = books[~(books["description"].isna()) &
       ~(books["num_pages"].isna()) &
       ~(books["average_rating"].isna()) &
       ~(books["published_year"].isna()) &
       ~(books["description"].isna()) ]

valid_books
```

```python
#category distribution
valid_books['categories'].value_counts().reset_index().sort_values('count', ascending=False)
```

```python
#count description length
valid_books['description_length'] = valid_books['description'].str.split().str.len()
valid_books
```

```python
valid_books.loc[valid_books['description_length'].between(25,34), 'description']
```

```python
# keep books with description word length 25 and more

valid_books_25_up = valid_books[valid_books['description_length'] >= 25]
valid_books_25_up
```

```python
# combine titles and subtitles if subtitles exist

valid_books_25_up['newtitle'] = (
    np.where(valid_books_25_up['subtitle'].isna(), valid_books_25_up['title'], 
    valid_books_25_up[["title", "subtitle"]].astype(str).agg(":| ".join,axis=1))
)
    
valid_books_25_up
```

```python
valid_books_25_up["tagged_description"] = valid_books_25_up[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
valid_books_25_up
```

```python
(
valid_books_25_up
.drop(["subtitle", "missing_description", "age_of_book", "description_length"], axis=1)
.to_csv("books_cleaned.csv", index = False)
)
```
